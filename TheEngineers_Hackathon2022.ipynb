## Description: This code reads all the dunNames, runs them in a loop and gets the URL using google search.

## Once the key words are captured, it will append the findings to a list and writes the findings into target file

## Step3. Main function which does a google search for a given dunName, goes through the html web content and searches for required keywords

```python
#Running Both again:


def get_diverse_owned(dunName):
    
    
    UrlList=[]
    for j in search(dunName, tld="co.in", num=10, stop=10, pause=2):
        UrlList.append(j)
    #print(UrlList[0])
    data = []
    woman=[]
    minority=[]
    try:
        page = urllib.request.urlopen(UrlList[0])
        print(f'This url is opening {UrlList[0]}')
        Correct_URLs.append(UrlList[0])
        please_break='dont break'
    except:
        try:
            page = urllib.request.urlopen(UrlList[0])
            print(f'This url is opening {UrlList[0]}')
            Correct_URLs.append(UrlList[0])
            please_break='dont break'
        except:
            print(f'Sorry. This url is not opening {UrlList[0]}')
            Problem_URLs.append(UrlList[0])
            woman_main.append(0)
            minority_main.append('NA')
            please_break='break'
        #raise Exception(f'Sorry. This url is not opening {UrlList[0]}')
    while please_break !='break':
        soup = BeautifulSoup(page, 'html.parser')
        ls = [x.get_text(strip=True) for x in soup.find_all("div")]
        data.append((ls))
        for i in range(0,len(data)):
            for match in data[i]:
                if ( ("woman" in match.lower()) or ("women" in match.lower()) or ("woman owned" in match.lower()) or ("women owned" in match.lower()) or ("woman-owned" in match.lower()) or ("women-owned" in match.lower()) or ('wbenc' in match.lower()) ):

                    #print("woman owned")
                    woman.append(1)
                if ("african" in match.lower()):
                    minority.append("African American")
                elif (("hispanic" in match.lower()) or ("latino" in match.lower())):
                    minority.append("Latino")
                elif (("native american" in match.lower()) or ("indigenous" in match.lower())):
                    minority.append("Native American")
                elif (("asian-paific" in match.lower()) or ("asian" in match.lower())):
                    minority.append("Asian-Paific American")
                elif (("asian-indian" in match.lower()) or ("indian" in match.lower())):
                    minority.append("Asian-Indian American")
                elif (("disabled" in match.lower()) or ("disabled owned" in match.lower())):
                    minority.append("Disabled Owned")
                elif (("lgbtq" in match.lower()) or ("lgbtq+ owned" in match.lower()) or ("nglcc" in match.lower()) or ("lgbtbe" in match.lower())):
                    minority.append("LGBTQIA+ Owned")
                elif (("veteran" in match.lower()) or ("veteran owned" in match.lower())):
                    minority.append("Veteran Owned")
                elif ("nmsdc" in match.lower()):
                    minority.append("Minority Owned")



            if len(woman)>=1:
                woman_main.append(1)
            else:
                woman_main.append(0)


            if len(minority)>=1:
                minority_main.append(minority[0])
            else:
                minority_main.append("NA")
            please_break='break'
```

## Step4. Call above function here in a loop

```python
#i=1
woman_main=[]
minority_main=[]
Problem_URLs=[]
Correct_URLs=[]
for dunsName in dunName_List:
    
    #get_diverse_owned(url,i)
    get_diverse_owned(dunsName)
   # i=i+1 
```

## Step5: Append the results to excel file under cols - isWomanOwned and MinorityOwnedDesc

```python
df['isWomanOwned']=woman_main
df['MinorityOwnedDesc']=minority_main
df.to_excel('Hackathon_Data_MinorityWomenOwned_2022 v1_solution.xlsx')
```
